### Getting the rest of municipalities

```python
muns=list(range(1,zm.shape[0]))

for i in muns:
    try:
        url=paste_url(zm.estado[i],zm.municipio[i])
        next_url,result_aux=scrapping_atr(url)

        while len(next_url)>0:
            print(next_url)
            time.sleep(1)
            next_url,aux=scrapping_atr(next_url)
            try:
                result_aux=result_aux.append(aux)
            except:
                print(f'The function didnt work with this url: /n {next_url}') 
        result=result.append(result_aux)
    except:
        print(f'The loop didnt work for {zm.municipio[i]}')

result.shape
```
